## data analysis for rep 2 

```python
df[df['rep'] == 2].describe()
```

```python
df2 = df[df['rep'] == 2]
df2.head()
```

```python
bays = df2['bay'].unique()
bays
```

```python
temp = {}
#lis -> ph mean, ph max, ph min, ec mean, ec max, ec min, temp max, temp min, plant height mean
for bay in bays:
    ph_mean = df2[df2['bay'] == bay]['pH'].mean()
    ph_max = df2[df2['bay'] == bay]['pH'].max()
    ph_min = df2[df2['bay'] == bay]['pH'].min()
    ec_mean = df2[df2['bay'] == bay]['EC'].mean()
    ec_max = df2[df2['bay'] == bay]['EC'].max()
    ec_min = df2[df2['bay'] == bay]['EC'].min()
    temp_mean = df2[df2['bay'] == bay]['h2o_temp_C'].mean()
    temp_max = df2[df2['bay'] == bay]['h2o_temp_C'].max()
    temp_min = df2[df2['bay'] == bay]['h2o_temp_C'].min()
    height_mean = df2[df2['bay'] == bay]['plant_height_cm'].mean()
    fresh_mass_mean = df2[df2['bay'] == bay]['harvest_fresh_mass_g'].mean()
    dry_mass_mean = df2[df2['bay'] == bay]['harvest_dry_mass_g'].mean()
    temp[bay] = [ph_mean, ph_max, ph_min,ec_mean, ec_max, ec_min,temp_mean, temp_max, temp_min, height_mean, fresh_mass_mean, dry_mass_mean]
temp
```

```python
for bay in bays:
    print(bay , df2[df2['bay'] == bay]['pH'].mean())
```

```python
for bay in bays:
    print(bay, df2[df2['bay'] == bay]['EC'].mean())
```

```python
rep2 = pd.DataFrame(data = temp)
rep2
```

```python
rep2 = rep2.transpose()
rep2
```

```python
rep2.reset_index(inplace=True)
rep2
```

```python
rep2.columns = ['bay','ph_mean','ph_max','ph_min','ec_mean','ec_max','ec_min','temp_mean','temp_max','temp_min','height_mean', 'fresh_mass_mean', 'dry_mass_mean']
rep2
```

```python
rep2.to_csv('rep2.csv')
```

```python
fig, ax = plt.subplots(figsize=(5,5))
sns.barplot(data=rep2, x = 'bay', y = 'height_mean', ax = ax).set(title='height_mean for all bay')
```

# analysis for all plants 

```python
df2.head()
```

```python
plants = df2['plant'].unique()
```

```python
plantdict = {}
for plant in plants:
    height_mean = df2[df2['plant'] == plant]['plant_height_cm'].mean()
    height_max = df2[df2['plant'] == plant]['plant_height_cm'].max()
    height_min = df2[df2['plant'] == plant]['plant_height_cm'].min()
    fresh_mass_mean = df2[df2['plant'] == plant]['harvest_fresh_mass_g'].mean()
    fresh_mass_max = df2[df2['plant'] == plant]['harvest_fresh_mass_g'].max()
    fresh_mass_min = df2[df2['plant'] == plant]['harvest_fresh_mass_g'].min()
    dry_mass_mean = df2[df2['plant'] == plant]['harvest_dry_mass_g'].mean()
    dry_mass_max = df2[df2['plant'] == plant]['harvest_dry_mass_g'].max()
    dry_mass_min = df2[df2['plant'] == plant]['harvest_dry_mass_g'].min()
    plantdict[plant] = [height_mean, height_max, height_min, fresh_mass_mean, fresh_mass_max, fresh_mass_min,dry_mass_mean,dry_mass_max, dry_mass_min]
plantdict
```

```python
rep2_plant = pd.DataFrame(data=plantdict)
rep2_plant
```

```python
rep2_plant = rep2_plant.transpose()
rep2_plant.reset_index(inplace=True)
col_name = ['plantNo','height_mean', 'height_max', 'height_min', 'fresh_mass_mean', 'fresh_mass_max', 'fresh_mass_min','dry_mass_mean','dry_mass_max', 'dry_mass_min']
rep2_plant.columns = col_name
rep2_plant.head(10)
```

```python
rep2_plant.describe()
```

```python
rep2_plant.to_csv('rep2_indiviual_plant.csv')
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data = rep2_plant, x = 'plantNo', y ='height_mean')
plt.title('mean height of plants')
plt.show()
```
